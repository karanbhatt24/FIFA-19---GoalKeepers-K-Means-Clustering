```python
data.plot(kind='scatter',x='GKScore',y='OverallPotential',color='red')
plt.show() 
# Selected attributes GKScore and OverallPotential because they both are on the same scale i.e out of 100 and
# they are interseting parameters to perform clustering, especially keeping football(soccer) tranfer market in mind
```

```python
# Segmenting above graph into different clusters using K-Means Clustering
X = data[['GKScore','OverallPotential']]
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42) #From the above graph we can see that apt value for
# number of clusters would 4
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
y = pd.DataFrame(y_kmeans)
```

```python
y.head()
```

```python
# To Visualize clustering
X = np.array(X)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of GK')
plt.xlabel('GKOverall')
plt.ylabel('OverallPtential')
plt.legend()
plt.show()
```
